# CLICKED ON AD PREDICTIVE MODEL - LOGISTIC REGRESSION

This project works with a false advertising data set, indicating whether a specific internet user clicked on an advertisement or not. As so the objective is to predict whether or not to click on an ad based on that userâ€™s features.

This dataset contains the following features:

* 'Daily Time Spent on Site': time on the site in minutes.
* 'Age': age of the consumer.
* 'Area Income': Average consumer income in the region.
* 'Daily Internet Usage': Average in minutes per day that the consumer is on the internet.
* 'Ad topic line': Ad title.
* 'City': City of the consumer.
* 'Male': Whether the consumer was male or not.
* 'Country': Country of the consumer.
* 'Timestamp': time when the consumer clicked on the ad or window closed.
* 'Clicked on Ad' ': 0 or 1 indicates whether or not you clicked on the ad.

## 1. Imports/Readings

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

#Output of plotting commands is displayed inline, directly below the code cell that produced it
%matplotlib inline
```

```python
#Read training data
ad_data = pd.read_csv('advertising.csv')
```

## 3. Predictive Model

```python
#Independent Variables
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]

#Variable target
y = ad_data['Clicked on Ad']

#Split the data to train/testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

#Implement the predictive model (logistic regression)
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)

#Apply the predictive model (logistic regression)
predictions = logmodel.predict(X_test)

#Return the classification report: 91% (could easily be better with further feature engineering)
print(classification_report(y_test,predictions))
```
